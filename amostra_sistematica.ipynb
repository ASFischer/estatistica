```python
import pandas as pd
import numpy as np
from math import ceil
```

```python
populacao = 150
amostra = 15 

k = ceil(populacao/amostra)
print(k)
#O calculo define que para ter 15 elementos de amostra devemos retirar 1 a cada 10
```

```python
r = np.random.randint (low= 1, high= k + 1, size = 1)
print(r)
```

```python
acumulador = r[0]
sorteados= []

for i in range(amostra): 
    sorteados.append(acumulador)
    acumulador += k
print(sorteados)
```

```python
df= pd.read_csv('iris.csv')
df_final = df.loc[sorteados]
df_final
```
